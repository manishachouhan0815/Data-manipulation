```python
import sqlite3
#connecting with the databases
con=sqlite3.connect('mydata2.sqlite')
#drop table if it already exist using execute method()
con.execute('drop table if exists test') #drop table if any present in database4.db
query="""
CREATE TABLE test 
(a VARCHAR(20), b VARCHAR(20), 
c REAL, d INTEGER
); """

con.execute(query)
con.commit()
```

```python
#insert data 
data=[('Atlanta', 'Georgia', 1.25, 6), ('Tallahassee', 'Florida', 2.6, 3), ('Sacramento', 'California', 1.7, 5)]
stmt= "INSERT INTO test VALUES(?,?,?,?)"  #? acts as a placeholder here
con.executemany(stmt,data) #executemany will exeute until the entire list is done
con.commit() #finally commiting everything 
```

```python
cursor=con.execute('select * from test') #select * from test state how many rows are there
rows= cursor.fetchall() #.fetchall will give the rows present
rows
```
