```python
#f['item_price'].astype('float')
df['item_price']=df['item_price'].str.replace(',','')
df['item_price']=df['item_price'].str.replace('$','')
df['item_price']
```

```python
dollar = lambda x: float(x[1:-1])
df.item_price = df.item_price.apply(dollar)
```

```python
#print a dataframe with only two columns item_name and item_price
```

```python
#1 delete the dupicates in item_name and quantity
filtered=df.drop_duplicates(["item_name","quantity"])
#2 select only the product with quantity equals to 1
one_prod= filtered[filtered.quantity==1]
#3 select only the item_name and item_price columns
price_per_item=one_prod[["item_name","item_price"]]

```

```python
#4 sort the values from the most to less expensive
price_per_item.sort_values(by=["item_name"])
```

```python
crime=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
crime.head()
```

```python
#convert the type of the column year to datetime
```

```python
crime.Year=pd.to_datetime(crime.Year, format='%Y')
crime.info()
                          
```

```python
# set the year column as the index of the dataframe
```

```python
crime=crime.set_index('Year', drop=True)
crime.head()
```

```python
#group the year by decades and sum the values 
```

```python
#use resample to sum each decade
crimes=crime.resample('10AS').sum()
# use resample to get the ,max value only for the "Population" column
population=crime["Population"].resample('10AS').max()
#updating the "POpulation" column
crimes["Population"]=population
crimes
```
